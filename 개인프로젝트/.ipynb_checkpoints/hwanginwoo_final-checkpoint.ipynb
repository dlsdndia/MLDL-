## 4. 모델 만들기

### 4 - 1) 모델에 쓰일 변수 선택

```python
var = ['id', 'route_id', 'vh_id', 'next_latitude', 'next_longitude', 'now_latitude',	'now_longitude',	'now_arrive_time',	'distance']
target = ['next_arrive_time']
```

```python
var = ['distance']
target = ['next_arrive_time']
```

### 4 - 2) 정규화

```python
df1 = data_train[var]
df1
```

```python
df2 = data_test[var]
df2
```

```python
df1_stats = df1.describe()
df1_stats = df1_stats.transpose()
df1_stats
```

```python
df2_stats = df2.describe()
df2_stats = df2_stats.transpose()
df2_stats
```

### 4 - 3) y( Label ) 분리

```python
y_train = np.asarray( data_train[target] )
```

```python
y_train
```

### 4 - 4) 정규화

```python
def normalization( x ):
  return ( x - df1_stats[ 'mean' ] ) / df1_stats[ 'std' ]
```

```python
def normalization( x ):
  return ( x - df2_stats[ 'mean' ] ) / df2_stats[ 'std' ]
```

```python
normed_train_df = normalization( df1 )
normed_test_df = normalization( df2 )
```

## 5. sftmax 회귀모델

### 5 - 1) 훈련 / 테스트 데이터 NumPy 배열로 변환

```python
X_train = np.asarray( normed_train_df ) 
X_test = np.asarray( normed_test_df )
```

```python
X_train
```

```python
y_train
```

### 5 - 2) Keras 모델구성

```python
X_train = np.array( X_train )
X_test = np.array( X_test )
```

```python
model = Sequential()
model.add( Dense( 1, input_dim = 1, activation = 'linear' ) )
sgd = optimizers.SGD( lr = 0.0001 )
model.compile( optimizer = sgd ,loss = 'mse', metrics = [ 'mse', 'mae' ] )
history = model.fit( X_train, y_train, batch_size = 100, epochs = 300, validation_split = 0.2 )
```

```python
history_dict = history.history
loss = history_dict[ 'loss' ]
val_loss = history_dict[ 'val_loss' ]
epochs = range( 1, len( loss ) + 1 )
```

```python
plt.plot( epochs, loss, 'b', label = 'Training loss' )
plt.plot( epochs, val_loss, 'r', label = 'Training val_loss' )
plt.title( 'Training and validation loss' )
plt.xlabel( 'Epochs' )
plt.ylabel( 'Loss' )
plt.legend()

plt.show()
```

```python
hist = pd.DataFrame(history.history)
hist[ 'epoch' ] = history.epoch
hist.tail()
```

### 5 - 3)  모델 평가

```python
model.evaluate( X_train, y_train )
```

# 5 - 4)  test 데이터를 기반으로 한 예측

```python
model.predict(X_test )
```
